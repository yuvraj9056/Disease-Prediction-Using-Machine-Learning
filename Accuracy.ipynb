```python
import numpy as np
import matplotlib.pyplot as plt
#accuracy
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
diabities = [0.7532467532467533, 0.7445887445887446, 0.6926406926406926, 0.7359307359307359]
Heart_disease = [0.8395061728395061, 0.8395061728395061, 0.7901234567901234, 0.8271604938271605]
Lung_cancer = [0.978494623655914, 0.956989247311828, 0.9354838709677419, 0.978494623655914]

# Set position of bar on X axis
br1 = np.arange(len(diabities))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, diabities, color ='r', width = barWidth,
		edgecolor ='grey', label ='diabities')
plt.bar(br2, Heart_disease, color ='g', width = barWidth,
		edgecolor ='grey', label ='heart_disease')
plt.bar(br3, Lung_cancer, color ='b', width = barWidth,
		edgecolor ='grey', label ='Lung_cancer')

# Adding Xticks
plt.xlabel('Models', fontweight ='bold', fontsize = 15)
plt.ylabel('Accuracy', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(diabities))],
		['SVM', 'naive_bayes', 'KNN', 'log_reg'])

plt.legend()
plt.show()
```
